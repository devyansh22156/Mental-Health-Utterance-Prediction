```python
import os
import re
import torch
import matplotlib.pyplot as plt  # For plotting loss curves
from torch.utils.data import Dataset, DataLoader
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    get_scheduler
)
from torch.optim import AdamW
from datasets import load_from_disk
from sentence_transformers import SentenceTransformer, util
from tqdm import tqdm
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score  # For METEOR score
from rouge_score import rouge_scorer
import bert_score
import nltk
nltk.download('wordnet')


# ------------------- Settings -------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "google/flan-t5-base"
MODEL_SAVE_PATH = "./finetuned_rag_flant5"
MAX_INPUT_LENGTH = 512
MAX_TARGET_LENGTH = 64
BATCH_SIZE = 8
EPOCHS = 6
LR = 5e-5
RETRIEVAL_TOP_K = 3

# ------------------- Load datasets -------------------
train_ds = load_from_disk("/content/drive/MyDrive/conv_data/therapy_train")
val_ds = load_from_disk("/content/drive/MyDrive/conv_data/therapy_val")
test_ds = load_from_disk("/content/drive/MyDrive/conv_data/therapy_test")

# ------------------- Tokenizer & Model -------------------
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)

if os.path.exists(MODEL_SAVE_PATH):
    print(f"Loading model from {MODEL_SAVE_PATH}")
    model = T5ForConditionalGeneration.from_pretrained(MODEL_SAVE_PATH).to(DEVICE)
else:
    model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME).to(DEVICE)

# ------------------- Retrieval Index -------------------
retrieval_model = SentenceTransformer("all-MiniLM-L6-v2").to(DEVICE)

def get_context(example):
    turns = example["input_text"].split("\n")
    return " ".join(turns[-2:]) if len(turns) >= 2 else example["input_text"]

train_contexts, train_responses = [], []
for ex in tqdm(train_ds, desc="Preparing Retrieval Index"):
    train_contexts.append(get_context(ex))
    train_responses.append(ex["target_text"])
train_embeddings = retrieval_model.encode(train_contexts, convert_to_tensor=True)

def retrieve_examples(query_context, k=RETRIEVAL_TOP_K):
    query_embedding = retrieval_model.encode(query_context, convert_to_tensor=True)
    cos_scores = util.cos_sim(query_embedding, train_embeddings)[0]
    top_k = torch.topk(cos_scores, k=min(k, len(train_contexts)))
    return [(train_contexts[i.item()], train_responses[i.item()]) for i in top_k.indices]

# ------------------- Dataset Class -------------------
class RetrievalAugmentedDataset(Dataset):
    def __init__(self, hf_dataset):
        self.data = hf_dataset

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        example = self.data[idx]
        context = get_context(example)
        retrieved = retrieve_examples(context)

        demo_text = ""
        for i, (ctx, resp) in enumerate(retrieved, start=1):
            demo_text += f"[Example {i}]\nContext: {ctx}\nResponse: {resp}\n\n"

        prompt = demo_text + f"Context: {context}\nPredict response:"
        input_ids = tokenizer(prompt, truncation=True, padding="max_length",
                              max_length=MAX_INPUT_LENGTH, return_tensors="pt").input_ids.squeeze()
        target_ids = tokenizer(example["target_text"], truncation=True, padding="max_length",
                               max_length=MAX_TARGET_LENGTH, return_tensors="pt").input_ids.squeeze()
        target_ids[target_ids == tokenizer.pad_token_id] = -100

        return {"input_ids": input_ids, "labels": target_ids}

# ------------------- Train -------------------
def train_model():
    model.train()
    train_loader = DataLoader(RetrievalAugmentedDataset(train_ds), batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(RetrievalAugmentedDataset(val_ds), batch_size=BATCH_SIZE)

    optimizer = AdamW(model.parameters(), lr=LR)
    scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=EPOCHS * len(train_loader))

    train_losses = []
    val_losses = []

    for epoch in range(EPOCHS):
        total_train_loss = 0
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1} Training"):
            input_ids = batch["input_ids"].to(DEVICE)
            labels = batch["labels"].to(DEVICE)

            outputs = model(input_ids=input_ids, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()

            total_train_loss += loss.item()

        avg_train_loss = total_train_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        print(f"[Epoch {epoch+1}] Training Loss: {avg_train_loss:.4f}")

        # Validation phase (set model to eval mode)
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f"Epoch {epoch+1} Validation"):
                input_ids = batch["input_ids"].to(DEVICE)
                labels = batch["labels"].to(DEVICE)

                outputs = model(input_ids=input_ids, labels=labels)
                loss = outputs.loss
                total_val_loss += loss.item()

        avg_val_loss = total_val_loss / len(val_loader)
        val_losses.append(avg_val_loss)
        print(f"[Epoch {epoch+1}] Validation Loss: {avg_val_loss:.4f}")

        # Set model back to train mode
        model.train()

    # Plot training and validation losses
    plt.figure(figsize=(8, 6))
    epochs = range(1, EPOCHS + 1)
    plt.plot(epochs, train_losses, label="Training Loss")
    plt.plot(epochs, val_losses, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Losses")
    plt.legend()
    plt.show()

    # Save model and tokenizer after training
    model.save_pretrained(MODEL_SAVE_PATH)
    tokenizer.save_pretrained(MODEL_SAVE_PATH)

# ------------------- Post-process predictions -------------------
def clean_generated_response(pred):
    response_parts = re.findall(r"Response:\s*(.*)", pred)
    if response_parts:
        final_response = response_parts[-1].strip()
        final_response = re.sub(r"\[SEP\].*", "", final_response).strip()
        return final_response
    return re.sub(r"\[SEP\].*", "", pred).strip()

# ------------------- Evaluation -------------------
# ------------------- Evaluation -------------------
def evaluate_model():
    model.eval()
    preds, refs = [], []
    smooth_fn = SmoothingFunction().method7
    # Initialize rouge scorer with rouge1, rouge2, and rougeL.
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    for example in tqdm(test_ds, desc="Evaluating"):
        context = get_context(example)
        retrieved = retrieve_examples(context, RETRIEVAL_TOP_K)

        demo_text = ""
        for i, (ctx, resp) in enumerate(retrieved, start=1):
            demo_text += f"[Example {i}]\nContext: {ctx}\nResponse: {resp}\n\n"

        prompt = demo_text + f"Context: {context}\nPredict response:"
        input_ids = tokenizer(prompt, return_tensors="pt", truncation=True,
                              max_length=MAX_INPUT_LENGTH, padding="max_length").input_ids.to(DEVICE)

        output = model.generate(input_ids=input_ids, max_length=MAX_TARGET_LENGTH)
        pred = tokenizer.decode(output[0], skip_special_tokens=True)
        pred_clean = clean_generated_response(pred)

        preds.append(pred_clean)
        refs.append(example["target_text"])

    # BLEU score
    bleu_scores = [
        sentence_bleu([ref.split()], pred.split(), smoothing_function=smooth_fn)
        for ref, pred in zip(refs, preds)
    ]
    avg_bleu = sum(bleu_scores) / len(bleu_scores)

    # ROUGE scores: F1 measures for rouge1, rouge2, and rougeL
    rouge1_scores = []
    rouge2_scores = []
    rougeL_scores = []
    for ref, pred in zip(refs, preds):
        scores = scorer.score(ref, pred)
        rouge1_scores.append(scores["rouge1"].fmeasure)
        rouge2_scores.append(scores["rouge2"].fmeasure)
        rougeL_scores.append(scores["rougeL"].fmeasure)
    avg_rouge1 = sum(rouge1_scores) / len(rouge1_scores)
    avg_rouge2 = sum(rouge2_scores) / len(rouge2_scores)
    avg_rougeL = sum(rougeL_scores) / len(rougeL_scores)

    # METEOR scores (per example)
    meteor_scores = []
    for ref, pred in zip(refs, preds):
        # Tokenize both reference and prediction since meteor_score expects pre-tokenized inputs.
        meteor_scores.append(meteor_score([ref.split()], pred.split()))
    avg_meteor = sum(meteor_scores) / len(meteor_scores)

    # BERTScore
    P, R, F1 = bert_score.score(preds, refs, lang="en", verbose=True)
    avg_bertscore = F1.mean().item()

    # Print scores
    print(f"\nBLEU: {avg_bleu:.4f}")
    print(f"ROUGE-1 F1: {avg_rouge1:.4f}")
    print(f"ROUGE-2 F1: {avg_rouge2:.4f}")
    print(f"ROUGE-L F1: {avg_rougeL:.4f}")
    print(f"METEOR: {avg_meteor:.4f}")
    print(f"BERTScore (F1): {avg_bertscore:.4f}")

    # Print sample predictions
    for i in range(min(3, len(preds))):
        print(f"\n=== Test Example {i+1} ===")
        print("Reference:", refs[i])
        print("Predicted:", preds[i])


# ------------------- Run -------------------
if __name__ == "__main__":
    if not os.path.exists(MODEL_SAVE_PATH):
        print("Training model...")
        train_model()
    else:
        print("Model already exists. Skipping training.")

    evaluate_model()
```
